```python
import matplotlib.pyplot as plt

def plot_district_precipitation(df, threshold_df, target_district, target_custom_year):
    """
    Plots daily precipitation for a specific district and custom year,
    overlaying the 10-year exceedance threshold.
    """
    # 1. Filter the daily data for the district and custom year
    # We sort by date to ensure the line plot connects correctly
    district_data = df[
        (df['DISTRICT'] == target_district) &
        (df['custom_year'] == target_custom_year)
    ].sort_values('date')

    if district_data.empty:
        print(f"No data found for {target_district} in custom year {target_custom_year}")
        return

    # 2. Get the threshold for this district
    # We use .item() to get the scalar value from the series
    try:
        threshold = threshold_df[threshold_df['DISTRICT'] == target_district]['THRESHOLD_10YR'].item()
    except (ValueError, IndexError):
        print(f"Threshold not found for {target_district}")
        return

    # 3. Plotting
    plt.figure(figsize=(12, 6))

    # Plot the daily precipitation
    plt.plot(district_data['date'], district_data['precip_max'],
             label='Daily Max Precip', color='steelblue', alpha=0.8)

    # Plot the threshold as a horizontal dashed line
    plt.axhline(y=threshold, color='red', linestyle='--', linewidth=2,
                label=f'10-Year Threshold ({threshold:.2f})')

    # Highlight the exceedance points (events)
    exceedances = district_data[district_data['precip_max'] > threshold]
    plt.scatter(exceedances['date'], exceedances['precip_max'],
                color='red', edgecolor='black', zorder=5, label='Exceedance Day')

    # Formatting
    plt.title(f"Precipitation Analysis: {target_district} ({target_custom_year}-{target_custom_year+1})")
    plt.xlabel("Date")
    plt.ylabel("Precipitation (mm/day)")
    plt.legend()
    plt.grid(True, linestyle=':', alpha=0.6)

    # Rotate dates for better readability
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Example Usage:
plot_district_precipitation(df, threshold_df, "PRAKASAM", 1994)
```
